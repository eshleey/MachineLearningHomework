```python
# Ses dosyasını yükleme
data, sr = librosa.load(dosya_yolu, sr=None) # sr=None, dosyanın orijinal örnekleme hızını korur
```

```python
# Teknik özellikleri hesaplama
sure_saniye = len(data) / sr
kanal_sayisi = 1 # librosa, varsayılan olarak sesi mono (tek kanal) yükler.
dtype = data.dtype

print("--- Ses Dosyasının Teknik Özellikleri ---")
print(f"1. Seçilen Dosya: {dosya_yolu}")
print(f"2. Örnekleme Oranı (sr): {sr} Hz")
print(f"3. Süre (saniye cinsinden): {sure_saniye:.3f} saniye")
print(f"4. Toplam Örnek Sayısı (Frame): {len(data)}")
print(f"5. Kanal Sayısı: {kanal_sayisi} (Mono)")
print(f"6. Veri Tipi (Bit Derinliği): {dtype}")
```
