```python
import matplotlib.pyplot as plt
import librosa.display
import numpy as np
```

```python
from google.colab import drive
drive.mount('/content/drive')
dosya_yolu = '/content/drive/MyDrive/UrbanSound8K/fold1/7061-6-0-0.wav'
data, sr = librosa.load(dosya_yolu, sr=None)
```

```python
# Grafiği çizdirme
plt.figure(figsize=(14, 5))
librosa.display.waveshow(y=data, sr=sr, x_axis='time')
plt.title(f'Zaman-Genlik Grafiği (Waveform) - Örnek Dosya')
plt.xlabel('Zaman (Saniye)')
plt.ylabel('Genlik (Amplitude)')
plt.grid(True)
plt.show()
```
